```python
import parselmouth 
from parselmouth import praat
import matplotlib.pyplot as plt
```

```python
test1 = '/home/tenet/Desktop/CS22Z121/Assignment_2/Task5/h.wav'
test2 = '/home/tenet/Desktop/CS22Z121/Assignment_2/Task5/swav.wav'
test3 = '/home/tenet/Desktop/CS22Z121/Assignment_2/a_16_1_c.wav'
```

```python
sound_ha = parselmouth.Sound(test1)
sound_sa = parselmouth.Sound(test2)
sound_a = parselmouth.Sound(test3)
```

```python
f0min=75
f0max=300
pointProcess = praat.call(sound_ha, "To PointProcess (periodic, cc)", f0min, f0max)
pointProcess = praat.call(sound_sa, "To PointProcess (periodic, cc)", f0min, f0max)
pointProcess = praat.call(sound_a, "To PointProcess (periodic, cc)", f0min, f0max)

```

```python
formants_ha = praat.call(sound_ha, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
formants_sa = praat.call(sound_sa, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
formants_a = praat.call(sound_a, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
```

```python
numPoints = praat.call(pointProcess, "Get number of points")
f1_ha = []
f2_ha = []
for point in range(0, numPoints):
    point += 1
    t = praat.call(pointProcess, "Get time from index", point)
    f1 = praat.call(formants_ha, "Get value at time", 1, t, 'Hertz', 'Linear')
    f2 = praat.call(formants_ha, "Get value at time", 2, t, 'Hertz', 'Linear')
    f1_ha.append(f1)
    f2_ha.append(f2)
```

```python
numPoints = praat.call(pointProcess, "Get number of points")
f1_sa = []
f2_sa = []
for point in range(0, numPoints):
    point += 1
    t = praat.call(pointProcess, "Get time from index", point)
    f1 = praat.call(formants_sa, "Get value at time", 1, t, 'Hertz', 'Linear')
    f2 = praat.call(formants_sa, "Get value at time", 2, t, 'Hertz', 'Linear')
    f1_sa.append(f1)
    f2_sa.append(f2)
```

```python
numPoints = praat.call(pointProcess, "Get number of points")
f1_a = []
f2_a = []
for point in range(0, numPoints):
    point += 1
    t = praat.call(pointProcess, "Get time from index", point)
    f1 = praat.call(formants_a, "Get value at time", 1, t, 'Hertz', 'Linear')
    f2 = praat.call(formants_a, "Get value at time", 2, t, 'Hertz', 'Linear')
    f1_a.append(f1)
    f2_a.append(f2)
```

```python
plt.scatter(f1_ha,f2_ha ,c="crimson")
#plt.scatter(f1_sa,f2_sa, c="darkorange")
plt.scatter(f1_a,f2_a ,  c="blue")
plt.xlabel("F1")
plt.ylabel("F2")
plt.title("Comparison of Formants for ha and a")
plt.legend(['ha', 'a'])

plt.show()
```
