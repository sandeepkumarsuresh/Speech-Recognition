### Q1. Generate two sinusoids at two different frequencies ω0 and ω1 (512 samples). Convolve the two signals .Compute the DFT of the convolved signal. Plot the spectrum. Compute the DFT of each of the signals. Multiply the spectra, and compare them with that of the convolved signal.Remember that the DFT order should be >= the length of the signal.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp



#Generating a Sine Wave at 50 Hz 
amp = 10
freq = 50
time = np.arange(0,1,1/512)
signal1 = amp*np.sin(2*np.pi*freq*time)

fig = plt.figure()
fig.set_size_inches(12,6)
plt.subplots_adjust(hspace=1)

plt.subplot(311)
plt.plot(time, signal1)
plt.title('Sine wave 50Hz')
plt.xlabel('Time [sec]')
plt.ylabel('Amplitude')
plt.xlim(0, 1)
plt.ylim(-15, 15)
```

```python
#Generating a sin wave at 25Hz
amp = 10
freq = 25
time = np.arange(0,1,1/512)
signal2 = amp*np.sin(2*np.pi*freq*time)

fig = plt.figure()
fig.set_size_inches(12,6)
plt.subplots_adjust(hspace=1)

plt.subplot(311)
plt.plot(time, signal2)
plt.title('Sine wave 25Hz')
plt.xlabel('Time [sec]')
plt.ylabel('Amplitude ')
plt.xlim(0, 1)
plt.ylim(-15, 15)
```

### Convolution of Signal 1 and Signal 2

```python
conv_signal = np.convolve(signal1,signal2,mode='full')
plt.plot(conv_signal )
plt.show()
```

### DFT of Convolved Signal

```python
dft_conv_signal =  np.fft.rfft(conv_signal)

magnitude = [np.sqrt(i.real**2+i.imag**2)/len(dft_conv_signal) for i in dft_conv_signal]
frequency = np.fft.rfftfreq(len(conv_signal),1/512)
plt.plot(frequency,magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("Dft of Convolved Signal")
plt.show()
```

### DFT of Signal 1 --> 50Hz

```python
dft_signal1 =  np.fft.rfft(signal1)
freq_signal1 = np.fft.rfftfreq(len(signal1),1/512)

plt.plot(freq_signal1,abs(dft_signal1))
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("Dft of Signal1 - 50hz")
plt.show()
```

### DFT of Signal 2 --> 25Hz

```python
dft_signal2 =  np.fft.rfft(signal2)
freq_signal2 = np.fft.rfftfreq(len(signal2),1/512)

plt.plot(freq_signal2,abs(dft_signal2))
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("Dft of Signal2 - 25hz")
plt.show()
```

### Multipying both the Spectrum and Plotting Them

```python
freq_signal3 =  dft_signal2*dft_signal1
plt.plot(abs(freq_signal3))
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("Multiplying the spectrum")
plt.show()
```
