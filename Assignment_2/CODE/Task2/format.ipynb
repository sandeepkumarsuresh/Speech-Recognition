```python
import parselmouth 
from parselmouth import praat
```

```python
test1 = '/home/tenet/Desktop/CS22Z121/Assignment_2/a_16_1_c.wav'
test2 = '/home/tenet/Desktop/CS22Z121/Assignment_2/e_16_1_c.wav'
test3 = '/home/tenet/Desktop/CS22Z121/Assignment_2/i_16_1_c.wav'
test4 = '/home/tenet/Desktop/CS22Z121/Assignment_2/o_16_1_c.wav'
test5 = '/home/tenet/Desktop/CS22Z121/Assignment_2/u_16_1_c.wav'
```

```python
sound_a = parselmouth.Sound(test1)
sound_e = parselmouth.Sound(test2)
sound_i = parselmouth.Sound(test3)
sound_o = parselmouth.Sound(test4)
sound_u = parselmouth.Sound(test5)
```

```python
f0min=75
f0max=300
pointProcess = praat.call(sound_a, "To PointProcess (periodic, cc)", f0min, f0max)
pointProcess = praat.call(sound_e, "To PointProcess (periodic, cc)", f0min, f0max)
pointProcess = praat.call(sound_i, "To PointProcess (periodic, cc)", f0min, f0max)
pointProcess = praat.call(sound_o, "To PointProcess (periodic, cc)", f0min, f0max)
pointProcess = praat.call(sound_u, "To PointProcess (periodic, cc)", f0min, f0max)
```

```python
formants_a = praat.call(sound_a, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
formants_e = praat.call(sound_e, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
formants_i = praat.call(sound_i, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
formants_o = praat.call(sound_o, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
formants_u = praat.call(sound_u, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
```

```python
numPoints = praat.call(pointProcess, "Get number of points")
f1_a = []
f2_a = []
f3_list = []
for point in range(0, numPoints):
    point += 1
    t = praat.call(pointProcess, "Get time from index", point)
    f1 = praat.call(formants_a, "Get value at time", 1, t, 'Hertz', 'Linear')
    f2 = praat.call(formants_a, "Get value at time", 2, t, 'Hertz', 'Linear')
    f3 = praat.call(formants_a, "Get value at time", 3, t, 'Hertz', 'Linear')
    f1_a.append(f1)
    f2_a.append(f2)
    f3_list.append(f3)
```
