```python
import parselmouth 
from parselmouth import praat
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
```

```python
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
# Code to get the Formant Countours

audio = parselmouth.Sound('a_16_1.wav')
f0min=75
f0max=300
pointProcess = praat.call(audio, "To PointProcess (periodic, cc)", f0min, f0max)
formants = praat.call(audio, "To Formant (burg)", 0.0025, 5, 5000, 0.025, 50)
numPoints = praat.call(pointProcess, "Get number of points")
f1_a = []
f2_a = []
time_a =[]
for point in range(0, numPoints):
    point += 1
    t = praat.call(pointProcess, "Get time from index", point)
    f1 = praat.call(formants, "Get value at time", 1, t, 'Hertz', 'Linear')
    f2 = praat.call(formants, "Get value at time", 2, t, 'Hertz', 'Linear')
    f3 = praat.call(formants, "Get value at time", 3, t, 'Hertz', 'Linear')
    f1_a.append(f1)
    f2_a.append(f2)
    time_a.append(t)

plt.scatter(time_a,f1_a)
plt.scatter(time_a,f2_a)
```
